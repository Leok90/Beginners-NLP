```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd
```

```python
def getPageLinks(pageRange):
    links = []
    for pageNo in range(pageRange):
        url = "https://series.naver.com/movie/recentList.nhn?orderType=sale&sortingType=&tagCode=&page=" + str(pageNo + 1)
        req = requests.get(url)
        soup = BeautifulSoup(req.text, 'lxml')
        movielinks = soup.select('div.lst_thum_wrap ul li a[href]')
        for movielink in movielinks:
            link = str(movielink.get('href'))
            links.append("https://series.naver.com"+link)
    return links
```

```python
getPageLinks(2)
```

```python
pages = []
for i in range(3):
    pages.extend(getPageLinks(i))
```

```python
pages
```

```python
titlelist=[]
contentlist=[]
imglist=[]
for i in range(len(pages)):
    url = pages[i]
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'lxml')
    content=soup.select('#content > div.end_dsc._close.NE\=a\:mvi')
    title=soup.select('#content > div.end_head.NE\=a\:mvi > h2')
    img = soup.select('#container > div.aside.NE\=a\:mvi > span.pic_area > img')
    if len(img)>0:
        img = img[0].get('src')
    if content and title and img:
        content =  content[0].get_text()
        content = content.replace('줄거리더보기','')
        content = content.strip()
        content = content.replace ('\r', ' ')
        title = title[0].get_text()
        title = title.replace('HD(1080)','')
        title = title.replace('자막','')
        title = title.strip()
        contentlist.append(content)
        titlelist.append(title)
        imglist.append(img)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
                {'title' :titlelist,
                 'content' : contentlist,
                 'imgurl' : imglist}
                )
```
