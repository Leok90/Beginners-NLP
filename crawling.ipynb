### 네이버 환율 정보 불러오기

```python
url = 'https://finance.naver.com/marketindex/exchangeList.nhn'
```

```python
req = requests.get(url)
```

```python
data = req.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
result = soup.select(".tit > a")
```

```python
result
```

```python
country = []
for r in result:
    # print(type(r))
    print(r.string.strip())
    country.append(r.string.strip())
```

```python
print(country)
```

```python
result2 = soup.select(".sale")
```

```python
result2
```

```python
selling = []
for r in result2:
    print(r.string)
    selling.append(r.string)
```

```python
selling
```

```python
data = list(zip(country, selling))
```

```python
print(data)
```
